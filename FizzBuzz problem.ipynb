# fizzbuzz problem 


# challenge:
    # print all the numbers from 1 to 100 with the following conditions.
    # 1- if the number is divisible by 3; print 'fizz'
    # 2- if the number is divisible by 3; print 'buzz'
    # 3- if the number is divisible by 3&5; print 'fizzbuzz'
    # 4- else just print the number

```python
# method 1
k=0
for i in range(1,101):
    if i%3==0 and i%5==0:
        print("fizzbuzz")
        k=k+1
    elif i%3==0:
        print ("fizz")
    elif i%5==0:
        print ("buzz")
    else:
        print (i)
print("number of time fizzbuzz is", k)       
```
